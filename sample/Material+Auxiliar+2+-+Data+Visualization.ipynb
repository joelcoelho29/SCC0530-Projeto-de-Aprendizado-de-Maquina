## Material Auxiliar 2: Data Visualization


Este exemplo tem como objetivo mostrar algumas técnicas de visualização de dados que auxiliam na tomada de decisão de um projeto de machine learning. (será utilizado o mesmo dataset do Material Auxiliar 1).

```python
''' Importa as bibliotecas Pandas e Numpy '''
import pandas as pd
import numpy as np
```

```python
''' Importa do sklearn o método que carrega o dataset breast cancer. Nesse dataset iremos classificar se uma 
pessoa tem um tumor benigno (valor de target igual a 1) ou maligno (valor de target igual a 0) no seio. 
'''
from sklearn.datasets import load_breast_cancer
```

```python
# dataset está no formato do sklearn
aux_dataset = load_breast_cancer()
# transforma o dataset para o formato pd.DataFrame
dataset = pd.DataFrame(aux_dataset['data'], columns = aux_dataset['feature_names'])
# inclui a classe
dataset['target'] = aux_dataset['target']
# mostra os exemplos do dataset
dataset
```

### 1. describe() 

.describe() é um método do pandas Data Frame que retorna diversas estatístiscas das colunas do dataframe. 

```python
dataset.describe()
```

### 2. Histograma

.hist() é um método do pandas Series e DataFrame que retorna o histograma de uma ou diversas colunas. O Histograma é muito utilizado para analisar a distribuição dos valores de um determinado atributo. 

```python
dataset[['mean radius', 'mean texture']].hist(bins=20)
```

### 3. Verificar balanceamento das classes

É de suma importância verificar se as classes estão balanceadas. Se há um desbalanceamento grande entre as classes por vezes é indicado utilizar técnicas de pré-processamento para balanceamento das classes, por outras vezes é indicado manter a proporção natural do problema. Porém, o balanceamento ou desbalanceamento de classes influencia qual medida será utilizada na avaliação do classificador, por exemplo a acurácia não é indicada para avaliar problemas onde as classes são desbalanceadas. 

```python
# para este plot utilizaremos a biblioteca seaborn, que é muito útil para gráficos no python
import seaborn as sns
# para plotar a proporção das classes, primeiro contamos os valores da coluna 'target' e transformamos em um frame
aux = pd.DataFrame(dataset['target'].value_counts())
# usamos o frame criado acima para plot a frequencia de cada classe no dataset
sns.barplot(x=aux.index, y='target', data=aux)
```

### 4. Histograma 2

Uma outra forma de usar o histograma é verificar a distribuição dos valores de um atributo separado pelo valor da classe. Assim, é possível identificar certos padrões (é o que os algoritmos de classifcação tentam fazer) para cada classe.

```python
# criar uma nova coluna com a descrição da classe
dataset['target by name'] = dataset['target'].replace({0 : 'malign', 1 : 'benign'})
# plota o histograme para as duas classes com o atributo 'mean radius'
sns.histplot(x='mean radius', hue='target by name', discrete=True,  stat='density', common_norm=False, 
             data=dataset)
```

### 5. Correlação

Gráfico de correlação é muito utilizado para determinar a correlação entre dois atributos (numéricos), caso haja uma alta correlação entre dois atributos é possível eliminar um dos dois atributos, pois é como se os dois atributos "tivessem a mesma informação", assim para diminuir a complexidade do problema um dos atributos é eliminado. Em tarefas de regressão a correlação de um atributo com a variável dependente (target) é utilizado para determinar a importância do atributo e também para seleção de atributos. 

```python
# tirar a correlação entre 3 atributos, senão ia ficar uma imagem muito grande
sns.pairplot(dataset[['mean radius', 'mean texture', 'mean perimeter']], corner=True)
```

Note que os atributos 'mean radius' e 'mean perimeter' tem uma alta correlação, já a 'mean perimeter' com 'mean texture' & 'mean radius' com 'mean texture' tem uma baixa correlação entre os atributos.
